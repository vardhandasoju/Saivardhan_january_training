```python
import pandas as pd
df = pd.read_csv("../data/processed/cleaned_data.csv")
df["Frequency"] = (
    df["NumWebPurchases"] +
    df["NumStorePurchases"] +
    df["NumCatalogPurchases"]
)
df["Monetary"] = (
    df["MntWines"] +
    df["MntFruits"] +
    df["MntMeatProducts"] +
    df["MntFishProducts"] +
    df["MntSweetProducts"] +
    df["MntGoldProds"]
)
df.rename(columns={"Recency":"Recency_days"}, inplace=True)
df.to_csv("../data/processed/feature_engineered_data.csv", index=False)
```
